```python
## helper functions

def df_columns(df):
    return list(df.columns)

def get_source(df):
    return df.name.split('_')[0]

def get_version(df, get_full):
    if get_full == False:
        return df.name.split('_')[1]
    elif get_full == True:
        return df.name

def get_meta(df):
    return meta_dict[get_version(df,True)]

def check_candidates(df):
    gdp_candidates_present = list(set(df.columns).intersection(set(gdp_candidates)))
    gdppc_candidates_present = list(set(df.columns).intersection(set(gdppc_candidates)))
    #pop_candidates_present = list(set(df.columns).intersection(set(pop_candidates)))

    #source = get_source(df)
    #if source == 'wdi':
    #    print('sanity check: gdppc can be null')

    if gdp_candidates_present == gdppc_candidates_present:
        #print('gdp_candidates_present')
        return gdp_candidates_present, [True, True]

    elif gdp_candidates_present != [] and gdppc_candidates_present == []:
        #print('gdp_candidates_present')
        return gdp_candidates_present, [True, False]

    elif gdp_candidates_present == [] and gdppc_candidates_present != []:
        #print('gdppc_candidates_present')
        return gdppc_candidates_present, [False, True]
  
    else:
        return ValueError 

    return gdp_candidates_present, gdppc_candidates_present #, pop_candidates_present

def check_gdp_type(df):
    df_name = df.name
    ## check meta dict
    gdp_type = meta_dict[df_name][1]

    if gdp_type == 'total':
        return True, 'total'
    elif gdp_type == 'pc':
        return False, 'pc'

def get_col_nums(df, col_lists):
    gdp_candidates_present = col_lists[0]
    gdppc_candidates_present = col_lists[1]
    pop_candidates_present = col_lists[2]

    gdp_col_loc = []
    gdppc_col_loc = []
    pop_col_loc = []

    for i in range(len(gdp_candidates_present)):
        gdp_col_loc.append(df.columns.get_loc(gdp_candidates_present[i]))

    for i in range(len(gdppc_candidates_present)):
        gdppc_col_loc.append(df.columns.get_loc(gdppc_candidates_present[i]))

    for i in range(len(pop_candidates_present)):
        pop_col_loc.append(df.columns.get_loc(pop_candidates_present[i]))
        
    return gdp_col_loc, gdppc_col_loc, pop_col_loc


def convert_relevant_column(df, convert_to):
    rel_col = check_candidates(df)[0][0]
    new_df = df.copy()
    df_name = df.name

    #new_df_name = df.name + '_converted'
    #new_df.name = new_df_name
    ## check meta_dict for column info 
    
    df_gdp_unit = meta_dict[df_name][2]

    if convert_to == df_gdp_unit:
        #print('Already in desired unit')
        return new_df, rel_col
    
    elif convert_to != df_gdp_unit:
        converted_column_name = rel_col + "_converted"
        #print(converted_column_name)
        if df_gdp_unit == 'total' and convert_to == 'millions':
            new_df[converted_column_name] = new_df[rel_col] / 1000000
        elif df_gdp_unit == 'millions' and convert_to == 'total':
            new_df[converted_column_name] = new_df[rel_col] * 1000000

    return new_df, converted_column_name

def create_standard_df(df, convert_to, version_full, drop_na_iso_nums):
    source = get_source(df)
    version = get_version(df, version_full)
    gdp_type = check_gdp_type(df)[0]
    priority = prio_dict[get_source(df)]
    base_year = get_meta(df)[0]
    #print(base_year)

    version_priority = version_prio_dict[get_version(df, True)]

    #base_year,gdp_type,gdp_unit = get_meta(df)
        
    #df_name = df.name

    new_df, rel_col = convert_relevant_column(df, convert_to)

    new_df = new_df.replace(',','',regex=True) ## remove any errant commas
    
    if drop_na_iso_nums == True:
        new_df = new_df[~new_df['iso_num'].isnull()]
        new_df['iso_num'] = new_df['iso_num'].astype('int')
    

    #new_df_name = df.name + '_standardised'

    #new_df.name = new_df_name

    
    standardised_column_names = ['iso_num', 'country', 'year', 'iso_num+year','base_year']
    standardised_column_names.append(rel_col)
    #print(standardised_column_names)

    new_df['iso_num+year'] = new_df['iso_num'].astype('str') + '_' + new_df['year'].astype('str') 
    new_df['base_year'] = base_year

    new_df = new_df[standardised_column_names]
    #print(gdp_type)
    #print(new_df.columns)
    if gdp_type == True:
        #if standardised_column_names[-1] in gdp_candidates:
        new_df.rename(columns={standardised_column_names[-1]:'gdp'},inplace=True)
        #new_df['gdp'] = new_df['gdp'].astype('int')
        #if drop_na_relevant_data == True:
        #    new_df = new_df[~(new_df['gdp'].isnull())]
        #else:
        #    print('oops')
    elif gdp_type == False:
        #if standardised_column_names[-1] in gdppc_candidates:
        new_df.rename(columns={standardised_column_names[-1]:'gdppc'},inplace=True)
        #new_df['gdppc'] = new_df['gdppc'].astype('int')
        #if drop_na_relevant_data == True:
        #    new_df = new_df[~(new_df['gdppc'].isnull())]
        #else:
        #   print('oops')

    
    
    new_df['source'] = source
    new_df['source_prio'] = priority
    new_df['version'] = version
    new_df['version_prio'] = version_priority

    new_df['priority_check'] = new_df['iso_num'].astype('str') + '_' + new_df['year'].astype('str') + '_' + new_df['source_prio'].astype('str')  + '_' + new_df['version_prio'].astype('str') 

    
    #new_df['base_year'] = base_year

    #new_df_name = df.name + '_standardised'

    #new_df.name = new_df_name
    #print(new_df.columns)

    #if drop_na_iso_nums == True:
        #new_df = new_df[~new_df['iso_num'].isnull()]

    # if incl_pop_data == True:
    #     trunc_combined_pop_dataset = combined_pop_dataset[['iso_num','year','country','pop']]
    #     trunc_combined_pop_dataset['pop_source'] = 'combined_pop_data_set'
    #     #trunc_combined_pop_dataset['iso_num+year'] = trunc_combined_pop_dataset['iso_num'].astype('str') + '_' + trunc_combined_pop_dataset['year'].astype('str') 
    #     new_df = pd.merge(new_df, trunc_combined_pop_dataset, on=['iso_num','year'], how='left')

    # if calculate_missing_gdp_or_gdppc == True:
    #     if gdp_type == True:
    #         new_df['gdppc'] = np.NaN
    #         new_df['gdppc'].fillna(new_df['gdp']/new_df['pop'],inplace = True)
    #     elif gdp_type == False:
    #         new_df['gdp'] = np.NaN
    #         new_df['gdp'].fillna(new_df['gdppc']*new_df['pop'],inplace = True)

    return new_df # source, version, new_df.name

def create_combined_df(convert_to, version_full, with_pop, drop_na_iso_nums,calculate_missing_gdp_and_gdppc,mark_duplicates):
    standardised_df_list = []
    for df in meta_list:
        #print(df.name)
        standardised_df_list.append(create_standard_df(df,convert_to,version_full,drop_na_iso_nums))

    combined_df = pd.concat(standardised_df_list)
    combined_df = combined_df.replace(',','',regex=True) ## remove any errant commas

    if with_pop == True:
        combined_df = merge_with_population(combined_df)

    if drop_na_iso_nums == True:
        combined_df = combined_df[~combined_df['iso_num'].isnull()]

    if calculate_missing_gdp_and_gdppc == True:
        combined_df = calculate_gdp_and_gdppc(combined_df)
    
    #if drop_na_gdppc == True:
    #    combined_df = combined_df[~combined_df['gdppc'].isnull()]
    
    #if drop_na_gdp == True:
    #    combined_df = combined_df[~combined_df['gdp'].isnull()]

    if mark_duplicates == True:
        combined_df = check_duplicates(combined_df)

    return combined_df

def merge_with_population(df):
    trunc_combined_pop_dataset = combined_pop_dataset[['iso_num','year','country','pop']]
    trunc_combined_pop_dataset['pop_source'] = 'combined_pop_data_set'
    #trunc_combined_pop_dataset['iso_num+year'] = trunc_combined_pop_dataset['iso_num'].astype('str') + '_' + trunc_combined_pop_dataset['year'].astype('str') 
    with_population = pd.merge(trunc_combined_pop_dataset,df, on=['iso_num','year'], how='outer')

    #with_population = with_population
    
    return with_population

def calculate_gdp_and_gdppc(df):
    calculate_gdppc = df[~(df['gdp'].isnull()) & ((df['gdppc'].isnull())) & ~(df['pop'].isnull())]
    calculate_gdp = df[(df['gdp'].isnull()) & ~((df['gdppc'].isnull())) & ~(df['pop'].isnull())]

    no_calc_100 = df[~(df['gdp'].isnull()) & ((df['gdppc'].isnull())) & (df['pop'].isnull())]
    no_calc_010 = df[(df['gdp'].isnull()) & ~((df['gdppc'].isnull())) & (df['pop'].isnull())]
    no_calc_001 = df[(df['gdp'].isnull()) & ((df['gdppc'].isnull())) & ~(df['pop'].isnull())]
    no_calc_000 = df[(df['gdp'].isnull()) & ((df['gdppc'].isnull())) & (df['pop'].isnull())]

    calculate_gdppc['gdppc'].fillna(calculate_gdppc['gdp']/calculate_gdppc['pop'],inplace = True)
    calculate_gdp['gdp'].fillna(calculate_gdp['gdppc'].astype('int')*calculate_gdp['pop'], inplace = True)

    calculated_df = pd.concat([calculate_gdppc,calculate_gdp,no_calc_100,no_calc_010,no_calc_001,no_calc_000])
    calculated_df = calculated_df.reset_index(drop=True)
    return calculated_df

def check_duplicates(df):
    temp_df = df.copy()
    temp_df['duplicate_row'] = df.duplicated(['iso_num','year'],keep=False)
    
    return temp_df

def show_duplicate_counts(df):
    return df.groupby('iso_num+year',as_index=False).size()


def prioritise(df, show_version_count, sanity_check):
    temp_df = df.copy()
    temp_df['duplicate_row'] = df.duplicated(['iso_num','year'],keep=False)
    #temp_df['priority_check'] = temp_df['iso_num'].astype('str') + '_' + temp_df['year'].astype('str') + '_' + temp_df['source_prio'].astype('str')  + '_' + temp_df['version_prio'].astype('str') 
    #temp_df['iso_num+year'] = temp_df['iso_num'].astype('str') + '_' + temp_df['year'].astype('str') 
    duplicate_rows_only = temp_df[temp_df['duplicate_row']==True].sort_values(by=['priority_check'],ascending=True,ignore_index=True)
    duplicate_rows_only = duplicate_rows_only.drop_duplicates(subset=['iso_num+year'],keep='first')
    no_duplicate_rows = temp_df[temp_df['duplicate_row']==False].sort_values(by=['priority_check'],ascending=True,ignore_index=True)

    after_dropping = pd.concat([duplicate_rows_only,no_duplicate_rows])

    if show_version_count == True:
        print(after_dropping['version'].value_counts())

    if sanity_check == True:
        iso_year_comb_no = len(set(temp_df['iso_num+year']))
        if len(set(after_dropping['iso_num+year'])) != iso_year_comb_no:
            print('Incorrect iso_num+year count in output')
        elif len(set(after_dropping['iso_num+year'])) == iso_year_comb_no:
            print('Correct iso_num+year count in output')

    return after_dropping
```

```python
def change_country_years_end_year(df, new_end_year):
    new = df.copy()
    old_end_year = max(df['end_year'])
    new['end_year'].replace(old_end_year,new_end_year,inplace=True)
    return new

def make_country_year_grid(country_years, check_all_iso_nums_in_output, include_iso_num_and_year_column):
    country_year_list = []
    for iso in set(country_years['iso_num']):
        idx = country_years[country_years['iso_num']==iso]['country'].index[0]
        country = country_years[country_years['iso_num']==iso]['country'][idx]
        start_year = country_years[country_years['iso_num']==iso]['start_year'][idx]
        end_year = country_years[country_years['iso_num']==iso]['end_year'][idx]
        region = country_years[country_years['iso_num']==iso]['region'][idx]
        row_list = []

    
        for i in range(start_year, end_year+1):
            new_row = [iso, country, i, region]
            row_list.append(new_row)
        country_year_list.append(pd.DataFrame(row_list,columns=['iso_num','country','year','region']))

        country_year_grid = pd.concat(country_year_list)

    if check_all_iso_nums_in_output == True:
        flag = set(country_year_grid['iso_num'])==set(country_years['iso_num'])
        if flag == True:
            print('All iso nums in output')
        elif flag == False:
            print('Not all iso nums in output.')
            print('Missing:', set(country_year_grid['iso_num']).difference(set(country_years['iso_num'])))
    
    if include_iso_num_and_year_column == True:
        country_year_grid['iso_num+year'] = country_year_grid['iso_num'].astype('str') + '_' + country_year_grid['year'].astype('str') 

    return country_year_grid
```
